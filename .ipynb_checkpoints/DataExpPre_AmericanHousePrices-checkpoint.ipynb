```python
import pandas as pd
import numpy as np

allData = pd.read_csv('American_Housing_Data_20231209.csv')
```

```python
# variable information 
allData.head()
```

```python
cols_to_check = ['City', 'State', 'Zip Code', 'County']
print(allData[cols_to_check].nunique())
```

```python
y = allData['Price'].copy()
X = allData.drop(columns=['City', 'State', 'Zip Code', 'Address', 'Price'])

X = pd.get_dummies(X, columns=['County'])
```

```python
X.head()
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np

df = pd.concat([X, y], axis=1)

columns = df.columns
num_cols = 1 
num_rows = math.ceil(len(columns) / num_cols)

plt.figure(figsize=(18, 4 * num_rows))

for i, col in enumerate(columns):
    plt.subplot(num_rows, num_cols, i + 1)
    
    is_numeric = pd.api.types.is_numeric_dtype(df[col])
    is_bool = pd.api.types.is_bool_dtype(df[col]) or (df[col].dtype == 'object' and df[col].nunique() <= 2)

    if is_numeric and not is_bool:
        plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
        plt.ylabel('Frekvencija')
    else:
        counts = df[col].value_counts()
        plt.bar(counts.index.astype(str), counts.values, color='lightgreen', edgecolor='black')
        plt.ylabel('Broj pojavljivanja')
        plt.xticks(rotation=45)
    
    plt.title(col, fontsize=12, fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
