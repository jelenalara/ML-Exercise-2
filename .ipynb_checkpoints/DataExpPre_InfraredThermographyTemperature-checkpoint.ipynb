```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
  
# fetch dataset 
infrared_thermography_temperature = fetch_ucirepo(id=925) 
  
# data (as pandas dataframes) 
X = infrared_thermography_temperature.data.features.copy()
y = infrared_thermography_temperature.data.targets.copy()
  
# variable information 
print(infrared_thermography_temperature.variables) 
```

```python
print(X['Age'].unique())
```

```python
#age preprocesing 
X['Age'] = X['Age'].replace({'21-25': '21-30', '26-30': '21-30'})
age_order = {'18-20': 0, '21-30': 1, '31-40': 2, '41-50': 3, '51-60': 4, '>60': 5}
X['AgeEncoded'] = X['Age'].map(age_order)
X = X.drop(columns=['Age'])
```

```python
#Ethnicity and gender one hot encoding 
X = pd.get_dummies(X, columns=['Ethnicity', 'Gender'])
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np

df = pd.concat([X, y], axis=1)

columns = df.columns
num_cols = 1 
num_rows = math.ceil(len(columns) / num_cols)

plt.figure(figsize=(18, 4 * num_rows))

for i, col in enumerate(columns):
    plt.subplot(num_rows, num_cols, i + 1)
    
    is_numeric = pd.api.types.is_numeric_dtype(df[col])
    is_bool = pd.api.types.is_bool_dtype(df[col]) or (df[col].dtype == 'object' and df[col].nunique() <= 2)

    if is_numeric and not is_bool:
        plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
        plt.ylabel('Frekvencija')
    else:
        counts = df[col].value_counts()
        plt.bar(counts.index.astype(str), counts.values, color='lightgreen', edgecolor='black')
        plt.ylabel('Broj pojavljivanja')
        plt.xticks(rotation=45)
    
    plt.title(col, fontsize=12, fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
