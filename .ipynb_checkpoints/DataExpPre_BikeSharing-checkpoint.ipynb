```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
  
# fetch dataset 
seoul_bike_sharing_demand = fetch_ucirepo(id=560) 
  
# data (as pandas dataframes) 
X = seoul_bike_sharing_demand.data.features 
y = seoul_bike_sharing_demand.data.targets 
  
# variable information 
print(seoul_bike_sharing_demand.variables) 
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import math

# 1. Spajanje feature-a i targeta u jedan DataFrame
df = pd.concat([X, y], axis=1)

columns = df.columns
num_cols = 1
num_rows = math.ceil(len(columns) / num_cols)

plt.figure(figsize=(18, 5 * num_rows))

for i, col in enumerate(columns):
    plt.subplot(num_rows, num_cols, i + 1)
    
    if pd.api.types.is_numeric_dtype(df[col]):
        plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
        plt.ylabel('Frekvencija')
    else:
        counts = df[col].value_counts()
        plt.bar(counts.index.astype(str), counts.values, color='lightgreen', edgecolor='black')
        plt.ylabel('Broj pojavljivanja')
    
    plt.title(col, fontsize=12, fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
print(y['Functioning Day'].value_counts(normalize=True) * 100)
```

```python
df_full = pd.concat([X, y], axis=1)

df_full = df_full[df_full['Functioning Day'] == 'Yes']

y = df_full['Rented Bike Count'] #new target 
X = df_full.drop(columns=['Rented Bike Count', 'Functioning Day'])
```

```python
#one hot encoding
X = pd.get_dummies(X, columns=['Seasons']) # One-Hot Encodin 'Seasons'

#string to integer 
X['Holiday'] = X['Holiday'].map({'No Holiday': 0, 'Holiday': 1})

#date transformation to month + day of the week 
X['Date'] = pd.to_datetime(X['Date'], dayfirst=True)
X['Month'] = X['Date'].dt.month
X['DayOfWeek'] = X['Date'].dt.dayofweek
X = X.drop(columns=['Date'])
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np

df = pd.concat([X, y], axis=1)

columns = df.columns
num_cols = 1 
num_rows = math.ceil(len(columns) / num_cols)

plt.figure(figsize=(18, 4 * num_rows))

for i, col in enumerate(columns):
    plt.subplot(num_rows, num_cols, i + 1)
    
    is_numeric = pd.api.types.is_numeric_dtype(df[col])
    is_bool = pd.api.types.is_bool_dtype(df[col]) or (df[col].dtype == 'object' and df[col].nunique() <= 2)

    if is_numeric and not is_bool:
        plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
        plt.ylabel('Frekvencija')
    else:
        counts = df[col].value_counts()
        plt.bar(counts.index.astype(str), counts.values, color='lightgreen', edgecolor='black')
        plt.ylabel('Broj pojavljivanja')
        plt.xticks(rotation=45)
    
    plt.title(col, fontsize=12, fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
